```python
from bs4 import BeautifulSoup
import csv
```

```python
HTMLfile = r"D:\beautifulsoup4\bs4\221\index.html"
with open(HTMLfile, 'r', encoding='utf-8') as content:
    soup = BeautifulSoup(content.read(), 'html')
```

```python
CSVfile_name = r"D:\beautifulsoup4\bs4\221\data.csv"
fields = ['Mã MH','Tên MH','Số tín chỉ',
        'Nhóm lớp','DK/ Sĩ số','Ngôn ngữ','Nhóm LT','Giảng viên','Nhóm BT','Giảng viên BT/TN','Sĩ số LT',
        'Thứ(Buổi 1)','Tiết(Buổi 1)','Phòng(Buổi 1)','CS(Buổi 1)',
        'Thứ(Buổi 2)','Tiết(Buổi 2)','Phòng(Buổi 2)','CS(Buổi 2)',
        'Thứ(Buổi 3)','Tiết(Buổi 3)','Phòng(Buổi 3)','CS(Buổi 3)',]
 
trs = soup.find_all('tbody')[20].find_all('tr', recursive=False)     
with open(CSVfile_name, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(fields)
    
    count = 0
    MaMH = TenMH = TinChi = ''

    
    for tr in trs:
        
        count_tr = 0

        if count % 2 == 1:
            tmp = tr.find_all('td', recursive=False)
            MaMH = tmp[2].contents[0]
            TenMH = tmp[3].contents[0]
            TinChi = tmp[4].contents[0]
            
        if count % 2 == 0 and count >= 2:
            
            tr_items = tr.find('table',class_='gridTable').find('tbody').find_all('tr',recursive=False)
            
            NhomLop = SiSo = NgonNgu = NhomLT = GiangVien = NhomBT = GiangVienBT = SiSoLT = ''

            
            for tr_item in tr_items:
                
                Thu1 = Tiet1 = Phong1 = CoSo1 = ''  
                Thu2 = Tiet2 = Phong2 = CoSo2 = ''  
                Thu3 = Tiet3 = Phong3 = CoSo3 = ''  
                
                if count_tr % 3 == 1:
                    temp = tr_item.find_all('td',recursive=False)
                    # (NhomLop,SiSo,NgonNgu,NhomLT,GiangVien,NhomBT,GiangVienBT,SiSoLT) = temp[0:8]
                    NhomLop = temp[0].contents[0].strip()
                    SiSo = temp[1].contents[0].strip()
                    NgonNgu = temp[2].contents[0].strip()
                    NhomLT = temp[3].contents[0].strip()
                    GiangVien = str(temp[4].contents[0])
                    NhomBT = temp[5].contents[0].strip()
                    GiangVienBT = str(temp[6].contents[0])
                    SiSoLT = temp[7].contents[0].strip()

                    if GiangVien == '<i>"Chưa/Đang phân công"</i>':
                        GiangVien = ''
                    if GiangVienBT == '<i>"Chưa/Đang phân công"</i>':
                        GiangVienBT = ''
                    
                if count_tr % 3 == 2:
                    temp = tr_item.find_all('td')[1].find('tbody').find_all('tr',recursive=False)
                    temp1 = temp[1].find_all('td',recursive=False)
                    
                    Thu1 = str(temp1[0].contents[0])
                    Tiet1 = temp1[1].contents[0].replace('-','').strip().replace('\n','').strip()
                    Phong1 = temp1[2].contents[0]
                    CoSo1 = temp1[3].contents[0]
                    
                    Tiet1 = ' '.join(Tiet1.split())
                    if Phong1 == '------ ':
                        Phong1 = ''
                    if Thu1 == 'Chưa biết ':
                        Thu1 = ''    
                    
                    
                    if len(temp) > 2:
                        temp2 = temp[2].find_all('td', recursive=False)
                        
                        Thu2 = str(temp2[0].contents[0])
                        Tiet2 = temp2[1].contents[0].replace('-','').strip().replace('\n',' ').strip()
                        Phong2 = temp2[2].contents[0]
                        CoSo2 = temp2[3].contents[0]
                    
                    Tiet2 = ' '.join(Tiet2.split())    
                    if Phong2 == '------ ':
                        Phong2 = ''
                    if Thu2 == 'Chưa biết ':
                        Thu2 = ''
                        
                    if len(temp) > 3:
                        temp3 = temp[3].find_all('td', recursive=False)
                        
                        Thu3 = str(temp3[0].contents[0])
                        Tiet3 = temp3[1].contents[0].replace('-','').strip().replace('\n',' ').strip()
                        Phong3 = temp3[2].contents[0]
                        CoSo3 = temp3[3].contents[0]
                    
                    Tiet3 = ' '.join(Tiet3.split())
                    if Phong3 == '------ ':
                        Phong3 = ''
                    if Thu3 == 'Chưa biết ':
                        Thu3 = ''
                    
                    writer.writerow([MaMH, TenMH, TinChi,
                                     NhomLop,SiSo,NgonNgu,NhomLT,GiangVien,NhomBT,GiangVienBT,SiSoLT,
                                     Thu1,Tiet1,Phong1,CoSo1,
                                     Thu2,Tiet2,Phong2,CoSo2,
                                     Thu3,Tiet3,Phong3,CoSo3,])
                count_tr+=1   
                
        count+=1
        
    
    
# print(count) 
print("Running successfully")
```
